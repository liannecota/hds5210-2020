# RegEx Filtering

```python
med_center_filter = data['Hospital Name'].str.contains('CENTER$')
med_centers = data[med_center_filter]
```

```python
med_centers.shape
```

```python
med_centers.head()
```

```python
hospital_filter = data['Hospital Name'].str.contains('HOSPITAL')
hospitals = data[hospital_filter]
```

```python
hospitals.shape
```

```python
hospitals.head()
```

```python
valid_addr = data['Address'].str.contains('[0-9]{1,6} [A-Z0-9 ]+')
```

```python
invalid = data[~valid_addr]
```
