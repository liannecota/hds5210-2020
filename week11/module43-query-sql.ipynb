# PandaSQL 

```python
import pandas as pd
import pandasql as ps
```

```python
data = pd.read_csv('complications.csv')
```

```python
data.head()
```

```python
sql = """
select City, count(distinct [Provider ID]) 
from data 
where State = 'CA' 
group by City
"""
```

```python
cities = ps.sqldf(sql, locals())
```

```python
cities
```

```python
cities[cities['City'] == 'LOS ANGELES']
```
